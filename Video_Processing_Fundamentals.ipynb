```python
import cv2

import numpy as np

import IPython.display as ipd

from moviepy.editor import *

from google.colab.patches import cv2_imshow
```

```python
video = cv2.VideoCapture('/content/bus_cif.y4m')
```

```python
fourcc = cv2.VideoWriter_fourcc(*'mp4v')
out_video = cv2.VideoWriter('output_video.mp4', fourcc, 30.0, (640, 360))
```

## Applying techniques like blurring, blurring, sharpening, thresholding, contrast stretching, increasing or decreasing intensity.

```python
while True:
    ret, frame = video.read()

    if ret:
        gray_frame = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)


        blurred_frame = cv2.GaussianBlur(gray_frame, (5,5), 0)
        alpha = 1.5
        beta = 0

        contrast_frame = cv2.convertScaleAbs(blurred_frame, alpha=alpha, beta=beta)


        out_video.write(cv2.cvtColor(contrast_frame, cv2.COLOR_GRAY2BGR))

        cv2_imshow(frame)
        cv2_imshow(contrast_frame)

        if cv2.waitKey(1) & 0xFF == ord('q'):
            break
    else:
        break

video.release()
out_video.release()
cv2.destroyAllWindows()
```
